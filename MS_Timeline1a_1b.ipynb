```python
#Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cbook as cbook
import matplotlib.dates as mdates
```

```python
#Reading CSV
us_and_nc_df = pd.read_csv("Resources/US_NC_Combined.csv")
us_and_nc_df.head()
```

```python
# Listing all th columns:
list(us_and_nc_df)
```

```python
#Copying original DF
positivecases_tests_df = us_and_nc_df.copy()
del positivecases_tests_df['Unnamed: 0']
list(positivecases_tests_df)
```

```python
#Cleaning DF
positivecases_tests_df = positivecases_tests_df.drop(columns = ['US DeathIncrease',
                                       'US inICUCumulative',
                                       'US inICUCurrently',
                                       'US HospitalizedIncrease',
                                       'US HospitalizedCurrently',
                                       'US HospitalizedCumulative',
                                       'US Negative',
                                       'US NegativeIncrease',
                                       'US OnVentilatorCumulative',
                                       'US OnVentilatorCurrently',
                                       'NC DeathIncrease',
                                       'NC HospitalizedCumulative',
                                       'NC HospitalizedCurrently',
                                       'NC HospitalizedIncrease',
                                       'NC InIcuCumulative',
                                       'NC InIcuCurrently',
                                       'NC Negative',
                                       'NC NegativeIncrease',
                                       'NC OnVentilatorCumulative',
                                       'NC OnVentilatorCurrently',])

list(positivecases_tests_df)
```

```python
#positivecases_tests_df = positivecases_tests_df.set_index("date")
```

```python
positivecases_tests_df
positivecases_tests_df.dropna()
```

```python
# Calculating ration of Positive Cases in NC vs USA
positivecases_tests_df["NC_positive_cases"] = (positivecases_tests_df["NC Positive"]/positivecases_tests_df["NC Population"])*100000
positivecases_tests_df["USA_positive_cases"] = (positivecases_tests_df["US Positive"]/positivecases_tests_df["US Population"])*100000
positivecases_tests_df["NC_positive_cases"].dropna()
positivecases_tests_df["USA_positive_cases"].dropna()
```

```python
#Formating Date for charting
positivecases_tests_df["date"] = pd.to_datetime(positivecases_tests_df["date"], format='%Y-%m-%d')
```

```python
#Saving the new DF
positivecases_tests_df.to_csv("Resources/positivecases_tests.csv", index = False)
```

```python
#Plotting the graphs

#POSITIVE CASES - NC X USA
plt.figure(1)
plt.plot(positivecases_tests_df["date"], positivecases_tests_df["USA_positive_cases"], color = "blue", label="US Positive Cases per Captita")
plt.plot(positivecases_tests_df["date"], positivecases_tests_df["NC_positive_cases"], color= "green", label ="NC Positive Cases per Captita")


# Create a legend for our chart
plt.legend(loc="best")
plt.title("COVID-19 Positive Cases per Capita")
plt.xlabel("Date")
plt.ylabel("Positive Cases per 100.000 people")
plt.xticks(fontsize="small")
plt.yticks(fontsize="small")
plt.show()
```

#### **Although NC and the USA have had similar behavior of positive cases of COVID-19, NC was always under the national ratio during the whole period.

```python
positivecases_tests_df["NC_testing"] = positivecases_tests_df["NC TotalTestResults"]/positivecases_tests_df["NC Population"]*100000
positivecases_tests_df["USA_testing"] = positivecases_tests_df["US TotalTestResults"]/positivecases_tests_df["US Population"]*100000
positivecases_tests_df["NC_testing"].dropna()
positivecases_tests_df["USA_testing"].dropna()
```

```python
# TESTS RATIO - NC X USA
plt.figure(2)
plt.plot(positivecases_tests_df["date"], positivecases_tests_df["USA_testing"], color = "blue", label="US Tests per Capita")
plt.plot(positivecases_tests_df["date"], positivecases_tests_df["NC_testing"], color= "green", label ="NC Tests per Capita")

# Create a legend for our chart
plt.legend(loc="best")
plt.title("COVID-19 Tests per Capita")
plt.xlabel("Date")
plt.ylabel("Tests per 100.000")
plt.xticks(fontsize="small")
plt.yticks(fontsize="small")
plt.show()
```

#### ** When we compare the test rates between NC and the USA, it is possible to identify the same trend. However, NC was under the national test rates.
